## Some examples of functions

1. Polynomial (Polynomials) :




### In Python, we define the above function as follows:

```python
    def f(x):
        return x**3 - 5*x**2 + 9
        
#  After the function is defined, we can test whether it is correct:
#     print f(3)
#     -9
#    print f(1)
#     5   
```

### drawing a function into a function graph can help us understand the changes in the function.



```python
def f(x):
    return x**3 - 5*x**2 + 9

import numpy as np
x = np.linspace(-5, 5, num = 100)
y = f(x)

import matplotlib.pyplot as plt
plt.plot(x,y)
plt.show()
    
    
```

## Exponential Functions





* In Python, the exponential function can be defined using the Euler constant as follows:


```python
import numpy as np
import matplotlib.pyplot as plt


def exp(x):
        return np.e**x
    
    
print(exp(1))    

# Or you can use numpy's own exponential function

print(np.exp(2))

x = np.linspace(-5, 5, num = 100)

plt.plot(x, exp(x))
plt.show()
```

## Logarithmic Functions







```python
    x = np.linspace(1,10,100,endpoint = False)
    y1 = np.log2(x)
    y2 = np.log(x)
    y3 = np.log10(x)
    plt.plot(x,y1,'red',x,y2,'yellow',x,y3,'blue')
    plt.show()
```

## Trigonometric functions

Sin ( -pi , to pi)

```python
# Sin wave

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-np.pi, np.pi, 201)

plt.plot(x, np.sin(x))

plt.xlabel('Angle [rad]')
plt.ylabel('sin(x)')
plt.show()
```
